### **Pertanyaan 1**
Berapa total peluang game dimenangkan **bidak putih** jika diketahui pemain menggunakan strategi pembuka **Sicilian Defense** atau **French Defense**?

---

- Langkah 1 <br>Menghitung probabilitas Game dimenangkan bidak putih dan French Defense digunakan sebagai strategi pembuka

```python
# Cari jumlah keseluruhan Game
n_game = len(game_catur)    # 0.5 point

# Kondisi Game dimenangkan oleh bidak putih - 1 point
pemenang_bidak_putih = game_catur['pemenang'] == 'white'    # 0.5 point

# Kondisi Game menggunakan strategi pembuka French Defense - 1 point
menggunakan_french_defense = game_catur['strategi_pembuka'] == 'French Defense'    # 0.5 point

# Cari game yang dimenangkan bidak putih dan French Defense digunakan sebagai strategi pembuka 
event_white_and_french = pemenang_bidak_putih & menggunakan_french_defense    # 0.5 point

# Cari Jumlah game yang dimenangkan bidak putih dan French Defense digunakan sebagai strategi pembuka 
n_event_white_and_french = len(game_catur[pemenang_bidak_putih])    # 0.5 point

# Cari probabilitas game yang dimenangkan bidak putih dan French Defense digunakan sebagai strategi pembuka 
P_white_french = n_event_white_and_french/n_game    # 0.5 point

print(f"Probabilitas French Defense digunakan sebagai strategi pembuka adalah {P_white_french:.2f}")
```

- Langkah 2 <br> Menghitung probabilitas Game dimenangkan bidak putih dan Sicilian Defense digunakan sebagai strategi pembuka

```python
# Cari jumlah keseluruhan Game
n_game = len(game_catur)   # 0.5 point

# Kondisi Game dimenangkan oleh bidak putih
pemenang_bidak_putih = game_catur['pemenang'] == 'white'    # 0.5 point

# Kondisi Game menggunakan strategi pembuka Sicilian Defense
menggunakan_sicilian_defense = game_catur['strategi_pembuka'] == 'Sicilian Defense'    # 0.5 point

# Cari game yang dimenangkan bidak putih dan Sicilian Defense digunakan sebagai strategi pembuka 
event_white_and_sicilian = pemenang_bidak_putih & menggunakan_sicilian_defense    # 0.5 point

# Cari Jumlah game yang dimenangkan bidak putih dan Sicilian Defense digunakan sebagai strategi pembuka 
n_event_white_and_sicilian = len(game_catur[event_white_and_sicilian])    # 0.5 point

# Cari probabilitas game yang dimenangkan bidak putih dan Sicilian Defense digunakan sebagai strategi pembuka 
P_white_sicilian = n_event_white_and_sicilian/n_game    # 0.5 point

print(f"Probabilitas Sicilian Defense digunakan sebagai strategi pembuka adalah {P_white_sicilian:.2f}")
```

- Langkah 3 <br> Menghitung total probabilitas kemenangan bidak putih yang menggunakan strategi French Defense dan Sicilian Defense

```python
# Menghitung total probabilitas kemenangan bidak putih yang menggunakan strategi French Defense dan Sicilian Defense

P_white_sicilian_french = P_white_french + P_white_sicilian   # 2 point

print(f"Total Probabilitas Game dimenangkan bidak putih dan strategi pembuka menggunakan Sicilian Defense dan French Defense adalah {P_white_sicilian_french:.2f}")
```

## **Case**: Purchase Mobil
> 3 pertanyaan, 20 point

- Sebuah dealer mobil mengumpulkan data customer dan calon customer dari divisi sales untuk mengetahui segmen customernya berdasarkan penghasilan
- Mereka ingin melihat probabilitas mobil terjual pada masing-masing segmen. 
- Selain itu mereka juga ingin mengefektifkan pelayanan terhadap customer potential.
- Berikut data yang didapatkan.

```python
cars_purchase = pd.read_csv("cars_purchase.csv")
cars_purchase.head()
```

Dalam data ini, hanya terdapat 2 kategori penghasilan customer yaitu penghasilan sedang atau penghasilan tinggi.

Hal ini dapat dilihat dari kategori yang ada di dalam kolom penghasilan

Calon customer yang datang ke dealer, ada yang membeli mobil dan ada juga yang tidak membeli mobil. Ini bisa dilihat dari informasi unik yang ada di kolom purchase

### **Pertanyaan 1**

Dari 1500 calon customer yang datang. Berapa probabilitas mobil dibeli oleh customer?

```python
# Silahkan tulis code anda disini
n_customer = len(cars_purchase)

# kondisi di mana mobil dibeli customer
car_purchased = cars_purchase['purchase'] == 'ya'

# banyaknya pelanggan yang membeli mobil
n_car_purchased = len(cars_purchase[car_purchased])

# peluang mobil dibeli customer
P_car_purchased = n_car_purchased/n_customer

print(f"Probabilitas mobil dibeli oleh customer: {P_car_purchased:.2f}")
```

### **Pertanyaan 2** 

Jika seorang calon customer datang dan diketahui penghasilannya termasuk ke kategori sedang. Berapa probabilitas calon customer tersebut akan membeli mobil?

```python
# Silahkan tulis code anda disini
# kondisi penghasilan customer
cust_income_medium = cars_purchase['penghasilan'] == 'sedang'

# irisan antara kondisi membeli dan penghasilan sedang
medium_income_purchased = cust_income_medium & car_purchased

# banyaknya customer berpenghasilan sedang yang membeli mobil
n_medium_purchaser = len(cars_purchase[medium_income_purchased])

# banyaknya customer dengan penghasilan sedang
n_medium = len(cars_purchase[cust_income_medium])
# conditiona probaility of purchasing given the customer belongs to medium income level
P_medium_purchase = n_medium_purchaser/n_medium

print(f"Probabilitas calon customer akan membeli mobil jika diketahui penghasilan sedang: {P_medium_purchase:.2f}")
```

### **Pertanyaan 3**

Jika calon customer yang datang termasuk ke segmen berpenghasilan tinggi, berapa probabilitas customer tersebut membeli mobil?

```python
# Silahkan tulis code anda disini
# kondisi customer berpenghasilan tinggi dan membeli
high_income = cars_purchase['penghasilan'] == 'tinggi'

# kondisi high income customer yang membeli
high_income_purchaser = high_income & car_purchased

# banyaknya event high income customer membeli mobil
n_high_purchaser = len(cars_purchase[high_income_purchaser])

# banyaknya event high income customer berkunjung
n_high_income = len(cars_purchase[high_income])

# conditional probability untuk customer membeli mobil jika diketahui berpenghasilan tinggi
P_high_purchase = n_high_purchaser/n_high_income

print(f"Peluang customer yang datang membeli mobil jika diketahui customer tersebut berpenghasilan tinggi: {P_high_purchase:.2f}")
```

## **Case**: Churn Analysis
> 7 pertanyaan, 17 point
---
- Churn adalah kondisi dimana pelanggan berhenti memakai produk atau berhenti berlangganan program
- Case ini bertujuan untuk **memahami kondisi churn dan membantu pengambilan keputusan banyak target pelanggan baru** di suatu perusahaan telekomunikasi
- Dirangkum lewat data berisi informasi id customer dan informasi churn (1) atau tidaknya (0) customer tersebut dalam 1 bulan terakhir.

Berikut adalah datanya:

```python
# Jalankan code ini untuk membaca data
# Catatan : CODE INI TIDAK BOLEH DIUBAH

def load_data(filename):
    # membaca data
    customer = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
    # mengambil kolom customerID dan Churn
    customer = customer[["customerID","Churn"]]
    # mengubah string menjadi numerik
    customer["Churn"] = [1 if (i=="Yes") else 0 for i in customer["Churn"]]
    return customer

customer = load_data("WA_Fn-UseC_-Telco-Customer-Churn.csv")
customer
```

---
Misal, kita **memilih secara acak seorang target pelanggan baru**,

### **Pertanyaan 1**
Berapa peluang seseorang tersebut akan churn?

---

```python
# Lengkapi code berikut

# ambil data customer churn
churn_condition = customer[customer['Churn']==1] 

# hitung banyak data customer churn
count_churn = len(churn_condition) 

# hitung banyak data keseluruhan customer
count_total_customer = len(customer) 

# hitung peluang pelanggan churn
p_churn = count_churn/count_total_customer

print(f"Peluang pelanggan baru acak akan churn sebesar {p_churn:.4f}")
```

### **Pertanyaan 2**
Berapa peluang 5 dari 10 pelanggan baru akan churn?

---

```python
# Lengkapi code berikut

# Hitung pmf distribusi binomial
p_churn_5_from_10 = binom.pmf(k = 5, n = 10, p = p_churn)

print(f"Peluang 5 dari 10 pelanggan baru acak akan churn sebesar {p_churn_5_from_10:.4f}")
```

### **Pertanyaan 3**
Berapa ekspektasi banyak pelanggan churn jika ada 100 pelanggan baru?

---

```python
# Lengkapi code berikut

# Hitung ekspected value binomial dengan n = 100 dan p = p_churn
expected_number_churn = 100 * p_churn

print(f"Ekspektasi banyak pelanggan churn dari 100 pelanggan baru adalah {round(expected_number_churn)}")
```

### **Pertanyaan 4**
Berapa rata rata rentang banyak customer churn dari ekspektasinya ketika ada 100 pelanggan baru?

---

```python
# Lengkapi code berikut

# Hitung varians binomial banyak pelanggan churn

variance_churn = 100**2 * p_churn - expected_number_churn

# Hitung standar deviasi banyak pelanggan churn
std_churn = np.power(variance_churn,0.5)

print(f"Rentang banyak pelanggan churn sekitar {round(expected_number_churn)}±{round(std_churn)} pelanggan")
```

### **Pertanyaan 5**
Berapa banyak target pelanggan baru yang harus didapatkan jika target pelanggan yang tidak churn adalah 100 orang?

---

```python
# Lengkapi code berikut

# Hitung nilai n yang menjadikan ekspected value tidak churn adalah 100 dan peluang tidak churn adalah 1-p_churn
expected_target_new_customer = 100
p_not_churn = 1-p_churn

n = expected_target_new_customer/p_not_churn

print(f"Target banyak pelanggan baru agar pelanggan yang tidak churn 100 orang adalah {round(n)}")
```

### **Pertanyaan 6** 
Berapa peluang saat mendapatkan pelanggan baru ke-5, pelanggan akan churn untuk pertama kali?

---

```python
# Lengkapi code berikut

# Hitung pmf distribusi yang bersangkutan
p_cust_until_first_churn = (1-p_not_churn)**4 * p_churn

print(f"Peluang saat mendapatkan pelanggan baru ke-5, pelanggan akan churn untuk pertama kali sebesar {p_cust_until_first_churn:.4f}")
```

### **Pertanyaan 7** 
Berapa ekspektasi banyak pelanggan baru sampai menemukan customer yang akan churn pertama kali?

---

```python
# Lengkapi code berikut

# Hitung ekspektasi distribusi yang bersangkutan
expected_number_cust_untill_first_churn = 1/p_churn

print(f"Ekspektasi banyak pelanggan baru sampai menemukan customer yang akan churn pertama kali adalah {round(expected_number_cust_untill_first_churn)}")
```
